```python
# Prepare data for modeling
X = df.drop('is_fraud', axis=1)
y = df['is_fraud']

# Standardize Features (important for distance-based models)
scaler = StandardScaler()

# Ensure all data in X is numeric before scaling
X_scaled = scaler.fit_transform(X)
```

```python
# Sample 1000 points for t-SNE to reduce computation (optional but common)
X_sample, _, y_sample, _ = train_test_split(X, y, train_size=1000, stratify=y, random_state=42)

# Scale the data (important for t-SNE)
scaler = StandardScaler()
X_sample_scaled = scaler.fit_transform(X_sample)

# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_sample_scaled)

# Create a DataFrame for plotting
df_tsne = pd.DataFrame({
    'TSNE1': X_tsne[:, 0],
    'TSNE2': X_tsne[:, 1],
    'Label': y_sample
})

# Map labels to names
df_tsne['Label'] = df_tsne['Label'].map({0: 'Not Fraud', 1: 'Fraud'})

# Plot
plt.figure(figsize=(8, 6))
scatter = sns.scatterplot(
    data=df_tsne,
    x='TSNE1',
    y='TSNE2',
    hue='Label',
    palette={'Not Fraud': 'blue', 'Fraud': 'red'},
    alpha=0.6
)

# Fix legend
handles, labels = scatter.get_legend_handles_labels()
plt.legend(handles=handles[1:], labels=labels[1:], title='Transaction Type')
plt.title('t-SNE Visualization of Transactions')
plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')
plt.grid(True)
plt.show()
```

```python

# Optionally, you can check the first few rows of the scaled data:
# Convert the scaled array back to a DataFrame with column names
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# Display the full table (formatted)
display(X_scaled_df)
```

```python
# Assuming you already have:
# X_scaled - your standardized features
# y - the target variable

# Step 1: Train-test split with stratify
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)

# Step 2: Apply SMOTE on the training data
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Step 3: Print class distribution before and after SMOTE
print("Class Distribution Before SMOTE:")
print(y_train.value_counts())

print("\nClass Distribution After SMOTE:")
print(pd.Series(y_train_smote).value_counts())
```

```python
# Create and train the Isolation Forest model
iso_model = IsolationForest(n_estimators=100, contamination='auto', random_state=42)
iso_model.fit(X_train_smote)

# Make predictions and convert to binary labels
iso_pred = iso_model.predict(X_test)
iso_pred_converted = [1 if val == -1 else 0 for val in iso_pred]  # -1 = anomaly  1

# Get confidence scores (negative distance from decision boundary)
confidence_scores = iso_model.decision_function(X_test) * -1

# Confusion Matrix (in table format)
conf_matrix = confusion_matrix(y_test, iso_pred_converted)
conf_df = pd.DataFrame(conf_matrix,
                       index=['Actual: Non-Fraud', 'Actual: Fraud'],
                       columns=['Predicted: Non-Fraud', 'Predicted: Fraud'])

print("\n Isolation Forest - Confusion Matrix:")
display(conf_df)

#  Classification Report (in table format)
class_report = classification_report(y_test, iso_pred_converted, output_dict=True)
class_report_df = pd.DataFrame(class_report).transpose()

print("\n Isolation Forest - Classification Report:")
display(class_report_df)

#  AUC-ROC Score
auc_score = roc_auc_score(y_test, confidence_scores)
print(f"\n Isolation Forest - AUC-ROC Score: {auc_score:.4f}")
```

```python

# Take a 1% stratified sample from X_train to reduce memory usage
X_sample_svm, _, y_sample_svm, _ = train_test_split(
    X_train, y_train, train_size=0.01, stratify=y_train, random_state=42
)

# Split into train/test for SVM
X_train_svm, X_test_svm, y_train_svm, y_test_svm = train_test_split(
    X_sample_svm, y_sample_svm, test_size=0.2, random_state=42
)

# Train One-Class SVM
one_class_svm = OneClassSVM(nu=0.1, kernel="rbf", gamma="scale")
one_class_svm.fit(X_train_svm)

# Predict and convert -1 to fraud (1), 1 to normal (0)
y_pred_svm = one_class_svm.predict(X_test_svm)
y_pred_svm_converted = [1 if x == -1 else 0 for x in y_pred_svm]

# Confusion Matrix
conf_matrix_svm = confusion_matrix(y_test_svm, y_pred_svm_converted)
conf_df_svm = pd.DataFrame(conf_matrix_svm,
                           index=['Actual: Non-Fraud', 'Actual: Fraud'],
                           columns=['Predicted: Non-Fraud', 'Predicted: Fraud'])

print("\n One-Class SVM - Confusion Matrix:")
display(conf_df_svm)

# Classification Report
class_report_svm = classification_report(y_test_svm, y_pred_svm_converted, output_dict=True)
class_report_df_svm = pd.DataFrame(class_report_svm).transpose()

print("\n One-Class SVM - Classification Report:")
display(class_report_df_svm)

# AUC-ROC Score
auc_score_svm = roc_auc_score(y_test_svm, y_pred_svm_converted)
print(f"\n One-Class SVM - AUC-ROC Score: {auc_score_svm:.4f}")
```

```python
# Define Autoencoder architecture
def build_autoencoder(input_dim):
    model = models.Sequential()
    model.add(layers.InputLayer(input_shape=(input_dim,)))
    model.add(layers.Dense(64, activation="relu"))
    model.add(layers.Dense(32, activation="relu"))
    model.add(layers.Dense(16, activation="relu"))
    model.add(layers.Dense(32, activation="relu"))
    model.add(layers.Dense(64, activation="relu"))
    model.add(layers.Dense(input_dim, activation="sigmoid"))
    model.compile(optimizer="adam", loss="mse")
    return model

# Build and train the model
autoencoder = build_autoencoder(X_train.shape[1])
autoencoder.fit(X_train, X_train, 
                epochs=50, 
                batch_size=32, 
                validation_data=(X_test, X_test),
                verbose=1)

# Predict and calculate reconstruction error
X_test_reconstructed = autoencoder.predict(X_test)
reconstruction_error = np.mean(np.abs(X_test - X_test_reconstructed), axis=1)

# Threshold (95th percentile  top 5% as fraud)
threshold = np.percentile(reconstruction_error, 95)
y_pred_autoencoder = [1 if e > threshold else 0 for e in reconstruction_error]

# Confusion Matrix
conf_matrix_ae = confusion_matrix(y_test, y_pred_autoencoder)
conf_df_ae = pd.DataFrame(conf_matrix_ae,
                          index=['Actual: Non-Fraud', 'Actual: Fraud'],
                          columns=['Predicted: Non-Fraud', 'Predicted: Fraud'])

print("\n Autoencoder - Confusion Matrix:")
display(conf_df_ae)

# Classification Report
class_report_ae = classification_report(y_test, y_pred_autoencoder, output_dict=True)
class_report_df_ae = pd.DataFrame(class_report_ae).transpose()

print("\n Autoencoder - Classification Report:")
display(class_report_df_ae)

# AUC-ROC Score
auc_score_ae = roc_auc_score(y_test, y_pred_autoencoder)
print(f"\ Autoencoder - AUC-ROC Score: {auc_score_ae:.4f}")
```

```python
# Reduce dataset size for faster testing
X_train_small = X_train[:10000]
y_train_small = y_train[:10000]
X_test_small = X_test[:3000]
y_test_small = y_test[:3000]

# Isolation Forest

iso_forest = IsolationForest(contamination=0.05, random_state=42)
iso_forest.fit(X_train_small)
y_pred_iso = iso_forest.predict(X_test_small)
y_pred_iso = [1 if x == 1 else 0 for x in y_pred_iso]


# One-Class SVM

svm = OneClassSVM(nu=0.1, kernel="rbf", gamma="scale")
svm.fit(X_train_small)
y_pred_svm = svm.predict(X_test_small)
y_pred_svm = [1 if x == 1 else 0 for x in y_pred_svm]


# Autoencoder

def build_autoencoder(input_dim):
    model = Sequential([
        layers.InputLayer(input_shape=(input_dim,)),
        layers.Dense(64, activation="relu"),
        layers.Dense(32, activation="relu"),
        layers.Dense(16, activation="relu"),
        layers.Dense(32, activation="relu"),
        layers.Dense(64, activation="relu"),
        layers.Dense(input_dim, activation="sigmoid")
    ])
    model.compile(optimizer="adam", loss="mse")
    return model

autoencoder = build_autoencoder(X_train_small.shape[1])
autoencoder.fit(
    X_train_small, X_train_small, 
    epochs=10, 
    batch_size=128, 
    validation_data=(X_test_small, X_test_small),
    verbose=1
)

# Predictions and anomaly scoring for Autoencoder
X_test_recon = autoencoder.predict(X_test_small)
recon_error = np.mean(np.abs(X_test_small - X_test_recon), axis=1)
threshold = np.percentile(recon_error, 95)
y_pred_auto = [1 if e > threshold else 0 for e in recon_error]

# Evaluation & Comparison

models = {
    "Isolation Forest": y_pred_iso,
    "One-Class SVM": y_pred_svm,
    "Autoencoder": y_pred_auto,
}

results = []

for name, preds in models.items():
    results.append({
        "Model": name,
        "AUC": roc_auc_score(y_test_small, preds),
        "Precision": precision_score(y_test_small, preds, zero_division=0),
        "Recall": recall_score(y_test_small, preds, zero_division=0),
        "F1-Score": f1_score(y_test_small, preds, zero_division=0)
    })

# Show comparison as formatted table
comparison_df = pd.DataFrame(results)

print("\nModel Comparison Summary (Fast Test Version):")
display(
    comparison_df.style
    .background_gradient(cmap="YlGnBu")
    .format({
        "AUC": "{:.4f}",
        "Precision": "{:.4f}",
        "Recall": "{:.4f}",
        "F1-Score": "{:.4f}"
    })
)
```
